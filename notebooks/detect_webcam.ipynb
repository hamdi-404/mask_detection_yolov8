```python
from ultralytics import YOLO
import cv2

# Load the trained model
model = YOLO("best.pt")  # path to your downloaded model

# Open webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Inference
    results = model.predict(source=frame, show=False, conf=0.5)

    # Plot results on the frame
    annotated_frame = results[0].plot()

    # Display the frame
    cv2.imshow("Webcam - Mask Detection", annotated_frame)

    # Exit on ESC key
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
